```python
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
paragraph = """Games are sometimes played purely for enjoyment, sometimes for achievement or reward as well. They can be played alone, in teams, or online; by amateurs or by professionals. The players may have an audience of non-players, such as when people are entertained by watching a chess championship. On the other hand, players in a game may constitute their own audience as they take their turn to play. Often, part of the entertainment for children playing a game is deciding who is part of their audience and who is a player. A toy and a game are not the same. Toys generally allow for unrestricted play whereas games present rules for the player to follow."""
```

```python
nltk.download('punkt')
```

```python
words=word_tokenize(paragraph)
```

```python
print(len(words))
```

```python
words
```

```python
#Empty list to store words
words_no_punc= []
#To Remove punctuation marks
for w in words:
    if w.isalpha():
      words_no_punc.append(w.lower())
```

```python
words_no_punc
```

```python
print
(len(words_no_punc))
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
#List stopwords
stopwords = set(stopwords.words('english'))
print(stopwords)
```

```python
new_words=[]
for w in words_no_punc:
    if w not in stopwords:
        new_words.append(w)
```

```python
print(new_words)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(new_words)
fdist.most_common(10)
```

```python
#Library
from wordcloud import WordCloud
```

```python
#Library to plot the wordcloud
import matplotlib.pyplot as plt
#Generating the worcloud
wordcloud = WordCloud().generate_from_frequencies(fdist)
```

```python
#Plot the wordcloud
plt.figure(figsize = (8,8))
plt.imshow(wordcloud)
```

```python
#Generating the worcloud
wordcloud = WordCloud().generate_from_frequencies(fdist)
#Plot the wordcloud
plt.figure(figsize = (8,8))
plt.imshow(wordcloud)
#To remove axis value
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
ship_image = np.array(Image.open("/content/drive/MyDrive/joystick.jpg"))
wordcloud = WordCloud().generate_from_frequencies(fdist)

wordcloud = WordCloud(
    background_color="white",
    mask=ship_image,
    stopwords=stopwords,
    contour_width=3,
    contour_color='steelblue',
    max_words=1000,
    max_font_size=10000,
    random_state=42
).generate_from_frequencies(fdist)

plt.figure(figsize=[10,10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
